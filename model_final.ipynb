# Gradient Boosting with CatBoost

```python
model_catboost = CatBoostClassifier(
            eval_metric='AUC:hints=skip_train~false',
            loss_function='Logloss',
            auto_class_weights='Balanced',
            verbose=100,
            iterations=500,
            depth = 3,
            l2_leaf_reg = 2.75,
            learning_rate=0.05,
            min_data_in_leaf = 1
)
```

```python
param = {
    'iterations' : [500],
    'depth' : [2, 3],
    'learning_rate' : [0.01, 0.05, 0.1],
    'l2_leaf_reg': np.linspace(2, 3.5, 3), 

    'min_data_in_leaf': [1, 3, 5]
}
model_cv = CatBoostClassifier(eval_metric='AUC:hints=skip_train~false',
                            loss_function='Logloss',
                            auto_class_weights='Balanced',
                            verbose=100)


grid_search = GridSearchCV(model_cv, param, cv=4, n_jobs=-1, scoring='roc_auc')
grid_search.fit(X, y)

grid_search.best_params_
```

{'depth': 3,

 'iterations': 500,

 'l2_leaf_reg': 2.75,

 'learning_rate': 0.05,
 
 'min_data_in_leaf': 1}

```python
cv_split = KFold(n_splits = 4)
cv_results = cross_validate(model_catboost, X_train, y_train, scoring='roc_auc', cv=cv_split, return_train_score=True)
cv_results
```

```python
cv_results['test_score'].mean()
```

```python
model_catboost.fit(X_train, y_train)
```

```python
proba = model_catboost.predict_proba(X_test)
```

```python
roc_auc_score(y_test, proba[:,1])
```

```python
lst_feature_importance = sorted(list(zip(X.columns, model_catboost.feature_importances_.tolist())), key=operator.itemgetter(1))
lst_feature_importance[1][1]
```

```python
unimportant_features = []
for element in lst_feature_importance:
    if element[1] == 0:
        unimportant_features.append(element[0])

len(unimportant_features)
```

# Ансамблевые методы

```python
stack_class = StackingClassifier(
                    estimators=[    ('Catboost', model_catboost),
                                    ('Random Forest', rand_f),
                                    ('LGBM', lgbm)],
                                    final_estimator=log
)
```
